2. Load the following packages: data.table, DT, and tidyverse.

```r
library(data.table)
library(DT)
library(tidyverse)
```

3. Import the credit data set.

```r
credit <- fread("Credit data.csv")
```

4. Create a data table of the raw credit data, followed by a data table of clean credit data.

```r
# Create data table of raw credit data
datatable(credit, options = list(scrollX = TRUE, searching = FALSE, pageLength = 5), caption = 'Table 1: Raw Auto Data')
# Create data table of clean credit data
credit.clean <- na.omit(credit)
datatable(credit.clean, options = list(scrollX = TRUE, searching = FALSE, pageLength = 5), caption = 'Table 2: Clean Auto Data')
```

5. Generate summary statistics.

```r
summary(credit.clean)
```

6. Create a histogram of age from the data set.

```r
# YOUR CODE HERE
hist(credit.clean$Age, 
     main="Histogram of Age",
    xlab="Age",
    col="#009ACD")
```

7. Create a bar chart of gender as a factor variable from the data set.

```r
# YOUR CODE HERE
credit.clean<-credit.clean[ , Gender :=as.factor(Gender)]

plot(credit.clean$Gender, 
     main="Bar Chart of Gender", 
     ylab= "Count", 
     xlab="Gender",
     ylim=c(0,200),
     col=c("#79CDCD","#9A32CD"),
    )





```

8. Create a scatterplot of age against gender from the data set. Gender must be transformed into a factor to create this scatterplot.

```r
# Create scatterplot
# YOUR CODE HERE
credit.clean<-credit.clean[ , Gender :=as.factor(Gender)]
plot(credit.clean$Age, credit.clean$Gender)
```

9. Create a box plot of age against gender from the data set.

```r
boxplot(Age ~ as.factor(Gender), data = credit.clean)
```
